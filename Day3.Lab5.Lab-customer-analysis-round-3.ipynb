```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
data.head()
```

```python
data._get_numeric_data()
data.select_dtypes(np.number)
```

### Show DataFrame info.

```python
data.info()
```

### Describe DataFrame.

```python
data.describe()
```

```python
data[data['Response'].isin(['Yes'])]
```

### Show a plot of the total number of responses.


```python
sns.displot(data['Response'])
plt.show()
```

```python
sns.countplot(x="Response", data=data)
plt.show()
```

### Show a plot of the response rate by the sales channel.

```python
sns.countplot(x="Response", hue="Sales Channel", data=data)
plt.show()
```

```python
crosstab = pd.crosstab(data['Response'], data['Sales Channel'])
```

### Show a plot of the response rate by the total claim amount.

```python
sns.boxplot(x="Response", y="Total Claim Amount", data=data)
plt.show()
```

### Show a plot of the response rate by income.

```python
sns.boxplot(x="Response", y="Income", data=data)
plt.show()
```
